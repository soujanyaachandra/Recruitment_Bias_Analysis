```python
import pandas as pd

# Load your dataset
data = pd.read_csv("job_applicant_dataset_trial4.csv")
print(data.head())  # Check the first few rows of the dataset
```

```python
from sklearn.preprocessing import LabelEncoder

# One-Hot Encoding for Gender and Race
gender_encoded = pd.get_dummies(data['Gender'], prefix='Gender')
race_encoded = pd.get_dummies(data['Race'], prefix='Race')

label_encoder = LabelEncoder()
data['Ethnicity'] = label_encoder.fit_transform(data['Ethnicity'])

# Concatenate the one-hot encoded columns with the original dataset
data_with_encoded = pd.concat([data, gender_encoded, race_encoded], axis=1)

# Drop the original categorical columns (Gender, Race)
data = data_with_encoded.drop(columns=['Gender', 'Race'])


# Label encoding for categorical variables (e.g., Gender, Race, Ethnicity)
# label_encoder = LabelEncoder()
# data['Gender'] = label_encoder.fit_transform(data['Gender'])
# data['Race'] = label_encoder.fit_transform(data['Race'])
# data['Ethnicity'] = label_encoder.fit_transform(data['Ethnicity'])

# Verify encoding
print(data.head())
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download necessary NLTK resources
import nltk

# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Function to clean text
def clean_text(text):
    text = re.sub(r'\W', ' ', text)  # Remove non-alphanumeric characters
    text = text.lower()  # Convert to lowercase
    text = ' '.join([lemmatizer.lemmatize(word) for word in text.split() if word not in stop_words])  # Lemmatize and remove stopwords
    return text

# Apply text preprocessing to 'Resume' and 'Job Description'
data['Resume'] = data['Resume'].apply(clean_text)
data['Job Description'] = data['Job Description'].apply(clean_text)

# Verify cleaned text
print(data[['Resume', 'Job Description']].head())
```

```python
from transformers import pipeline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the sentiment analysis pipeline
sentiment_analyzer = pipeline('sentiment-analysis')

# If dataset is large, limit the rows for quick testing (for example, first 1000 rows)
data_sample = data.head(1000)

# Create a batch for sentiment analysis
batch_size = 16  # Adjust batch size for efficiency
sentiments = []

# Process job descriptions in batches
for i in range(0, len(data_sample), batch_size):
    batch = data_sample['Job Description'].iloc[i:i+batch_size].tolist()
    sentiment_results = sentiment_analyzer(batch)  # Process the batch at once
    sentiments.extend([result['label'] for result in sentiment_results])  # Append results

# Add sentiment results to the DataFrame
data_sample['Sentiment'] = sentiments

# Visualize sentiment distribution
sns.countplot(x='Sentiment', data=data_sample)
plt.title('Sentiment Distribution of Job Descriptions')
plt.show()
```

```python
# Example: Detecting gendered words in job descriptions using a simple list of masculine and feminine terms
masculine_words = ['strong', 'leader', 'aggressive', 'competitive']
feminine_words = ['nurturing', 'compassionate', 'supportive', 'cooperative']

def detect_gendered_language(text):
    masculine_count = sum([word in text for word in masculine_words])
    feminine_count = sum([word in text for word in feminine_words])
    return masculine_count, feminine_count

# Apply detection function
data[['Masculine Count', 'Feminine Count']] = data['Job Description'].apply(detect_gendered_language).apply(pd.Series)

print(data['Masculine Count'])

# Visualize results
sns.histplot(data['Masculine Count'], kde=True, label='Masculine Words')
sns.histplot(data['Feminine Count'], kde=True, label='Feminine Words')
plt.legend()
plt.title('Distribution of Masculine and Feminine Words in Job Descriptions')
plt.show()
```

```python
from gensim.models import KeyedVectors
import numpy as np
import pandas as pd

# Load Word2Vec pre-trained embeddings (download separately if not already done)
word2vec_path = 'GoogleNews-vectors-negative300.bin'
word_vectors = KeyedVectors.load_word2vec_format(word2vec_path, binary=True)
```

```python
def get_embedding(text, word_vectors):
    words = text.split()
    embeddings = [word_vectors[word] for word in words if word in word_vectors]
    if embeddings:
        return np.mean(embeddings, axis=0)
    else:
        return np.zeros(word_vectors.vector_size)

def cosine_similarity(vec1, vec2):
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))
```

```python
data['Job_Embedding'] = data['Job Description'].apply(lambda desc: get_embedding(desc, word_vectors))
```

```python
# Define gender terms
male_terms = ['he', 'male', 'man', 'him', 'his']
female_terms = ['she', 'female', 'woman', 'her', 'hers']

# Get embeddings for gender terms
male_embeddings = [word_vectors[word] for word in male_terms if word in word_vectors]
female_embeddings = [word_vectors[word] for word in female_terms if word in word_vectors]

def weat_score(job_embeddings, male_embeddings, female_embeddings):
    weat_scores = []
    for job_emb in job_embeddings:
        male_similarity = np.mean([cosine_similarity(job_emb, male_emb) for male_emb in male_embeddings])
        female_similarity = np.mean([cosine_similarity(job_emb, female_emb) for female_emb in female_embeddings])
        weat_scores.append(male_similarity - female_similarity)
    return weat_scores

# Calculate WEAT scores for all job descriptions
data['WEAT_Score'] = weat_score(data['Job_Embedding'], male_embeddings, female_embeddings)
```
