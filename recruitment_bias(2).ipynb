```python
import numpy as np

# List of columns to transform
columns_to_transform = ['Job Applicant Name', 'Ethnicity','Job Roles', 'Race_Mongoloid/Asian', 'Race_Negroid/Black','Race_White/Caucasian']

for column in columns_to_transform:
    # Check for negative or zero values
    min_value = main_df[column].min()

    # Shift the data to make sure all values are positive (if the data contains 0 or negative values)
    if min_value <= 0:
        main_df[column] = main_df[column] + abs(min_value) + 1  # Shift to positive

    # Apply power transformation
    main_df[column] = np.power(main_df[column], 0.25)

    # Check if there are any NaN values and handle them
#     print(f"NaN values in {column}: {main_df[column].isna().sum()}")  # Print the count of NaN values
    main_df[column].fillna(main_df[column].mean(), inplace=True)  # Replace NaN values with the column mean

# Check skewness again for all columns
print(main_df.skew())
```

```python
from sklearn.model_selection import train_test_split

# Assuming main_df is your DataFrame and 'Best Match' is the target column

# Step 1: Prepare the data (X - Features, y - Target)
X = main_df.drop('Best Match', axis=1)  # Drop the target column 'Best Match'
y = main_df['Best Match']  # Target variable (binary)

# Step 2: Split the data into training and testing sets (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Print the training and testing datasets
print("Training Features (X_train):")
print(X_train)  # Print first few rows of the training features
print("\nTesting Features (X_test):")
print(X_test)  # Print first few rows of the testing features

print("\nTraining Target (y_train):")
print(y_train)  # Print first few rows of the training target variable
print("\nTesting Target (y_test):")
print(y_test)  # Print first few rows of the testing target variable
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Step 4: Train the Logistic Regression model
logreg = LogisticRegression(random_state=42)
logreg.fit(X_train, y_train)

# Step 5: Make predictions on the test set
y_pred = logreg.predict(X_test)

# Step 6: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Step 4: Train the Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)  # 100 trees in the forest
rf_classifier.fit(X_train, y_train)

# Step 5: Make predictions on the test set
y_pred = rf_classifier.predict(X_test)

# Step 6: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
