```python
import matplotlib.pyplot as plt
import seaborn as sns

# Reusable plotting function with BERT-specific title
def plot_weat_group(results, title, figsize=(10, 4), palette="viridis"):
    sorted_results = dict(sorted(results.items(), key=lambda x: x[1]))

    plt.figure(figsize=figsize)
    sns.set_style("whitegrid")
    sns.barplot(
        x=list(sorted_results.values()),
        y=list(sorted_results.keys()),
        palette=palette,
        orient="h"
    )

    # Add value labels on bars
    for i, (label, value) in enumerate(sorted_results.items()):
        plt.text(value + 0.02 * (1 if value > 0 else -1), i,
                 f"{value:.3f}", va='center', fontsize=10)

    plt.axvline(0, color="gray", linestyle="--")
    plt.xlabel("WEAT Effect Size")
    plt.title(f"{title} (BERT Model)")
    plt.tight_layout()
    plt.show()


# -----------------------------
#  GROUPED WEAT SCORES (BERT)
# -----------------------------

gender_bias = {
    "STEM vs. Male": -0.2813,
    "STEM vs. Female": 0.2813,
    "Non-STEM vs. Male": -0.4034,
    "Non-STEM vs. Female": 0.4034,
}

first_name_bias = {
    "White vs. Black": 0.5225,
    "White vs. Asian": 0.9938,
    "Black vs. Asian": 0.4713,
}

stem_ethnicity_bias = {
    "White vs. Black": 1.4517,
    "White vs. Asian": 2.9000,
    "Black vs. Asian": 1.4483,
}

stem_best_match_bias = {
    "White vs. Black": 0.5230,
    "White vs. Asian": 0.9406,
    "Black vs. Asian": 0.4176,
}

age_bias = {
    "STEM vs. Non-STEM": 0.0102
}

# -----------------------------
#  PLOTTING ALL (BERT)
# -----------------------------

plot_weat_group(gender_bias, "Gender Bias", palette="coolwarm")
plot_weat_group(first_name_bias, "First Name Bias (Ethnicity)", palette="mako")
plot_weat_group(stem_ethnicity_bias, "STEM Bias by Ethnicity", palette="rocket")
plot_weat_group(stem_best_match_bias, "STEM Best Match Bias by Ethnicity", palette="crest")
plot_weat_group(age_bias, "Age Bias", palette="flare")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Put gender bias comparison data into a DataFrame
gender_bias_comparison = {
    "STEM vs Male": {"BERT": -0.2813, "GPT-2": -0.2995, "GPT-Neo": -0.1427},
    "STEM vs Female": {"BERT": 0.2813, "GPT-2": 0.2995, "GPT-Neo": 0.1427},
    "Non-STEM vs Male": {"BERT": -0.4034, "GPT-2": -0.3458, "GPT-Neo": -0.2225},
    "Non-STEM vs Female": {"BERT": 0.4034, "GPT-2": 0.3458, "GPT-Neo": 0.2225}
}

df = pd.DataFrame(gender_bias_comparison).T.reset_index().melt(id_vars="index")
df.columns = ["Bias Type", "Model", "Effect Size"]

# 2. Plot grouped bar chart
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="Bias Type", y="Effect Size", hue="Model", palette="Set2")
plt.axhline(0, color="gray", linestyle="--")
plt.xticks(rotation=45)
plt.title("Gender Bias Comparison Across Models (WEAT Effect Size)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# First Name Bias Comparison Data
first_name_bias_comparison = {
    "White vs Black": {
        "BERT": 0.5225,
        "GPT-2": -0.2905,
        "GPT-Neo": -0.3927
    },
    "White vs Asian": {
        "BERT": 0.9938,
        "GPT-2": -0.3111,
        "GPT-Neo": -0.3713
    },
    "Black vs Asian": {
        "BERT": 0.4713,
        "GPT-2": -0.0206,
        "GPT-Neo": 0.0214
    }
}

# Convert to DataFrame
df_name = pd.DataFrame(first_name_bias_comparison).T.reset_index().melt(id_vars="index")
df_name.columns = ["Bias Type", "Model", "Effect Size"]

# Plotting
plt.figure(figsize=(10, 5))
sns.set_style("whitegrid")
sns.barplot(data=df_name, x="Bias Type", y="Effect Size", hue="Model", palette="Set2")

# Add reference line at zero
plt.axhline(0, color="gray", linestyle="--")

# Add annotations
for i in range(len(df_name)):
    value = df_name.iloc[i]['Effect Size']
    if abs(value) > 0.01:
        plt.text(
            x=i % 3 - 0.3, y=value + 0.02 * (1 if value > 0 else -1),
            s=f"{value:.2f}", fontsize=9, ha='center'
        )

plt.title("First Name Bias Comparison Across Models (WEAT Effect Size)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
